```python
a = requests.get("https://api.cryptowat.ch/markets/bitflyer/btcjpy/ohlc")
```

```python
arr = array(a.json()["result"]['14400'])
```

```python
fig, ax = plt.subplots()
x = arr[:,0]
o = arr[:,1]
h = arr[:,2]
l = arr[:,3]
c = arr[:,4]
candlestick2_ohlc(ax, o, h, l, c)
xdate = [datetime.fromtimestamp(i) for i in x]
ax.xaxis.set_major_locator(ticker.MaxNLocator(6))

def mydate(x, pos):
    try:
        return xdate[int(x)]
    except IndexError:
        return ''

ax.xaxis.set_major_formatter(ticker.FuncFormatter(mydate))
#ax.format_xdata = mdates.DateFormatter('%Y-%m-%d')

fig.autofmt_xdate()
fig.tight_layout()
```
